```python
%reload_ext autoreload
%autoreload 2

import numpy as np
import jax.numpy as jnp
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
import seaborn as sns
```

```python
x = np.arange(0, 10, .1)

a = 1
b = 1

L = 1
H = 1
v = 1

color = iter(cm.rainbow(np.linspace(0, 1, 3)))

plt.figure(figsize=(6, 4))

for a in [1, 2, 3]:
    c = next(color)
    y = L + jnp.maximum(
        0, -1 + (1 + H) / jnp.power(1 + (jnp.power(1 + H, v) - 1) * jnp.exp(-b * (x - a)), 1 / v)
    )

    sns.lineplot(x=x, y=y, color=c, label=f"$a={a}$")

plt.xticks(np.arange(0, 11, 1))
plt.title(f"Effect of varying parameter $a$. All other parameters are 1.")
plt.xlabel("$x$")
plt.ylabel("$y(x)$")
plt.legend(loc="lower right")
```
