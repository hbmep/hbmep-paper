```python
%reload_ext autoreload
%autoreload 2

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
def get_scores(df: pd.DataFrame, metric="mae"):
    run = sorted(df.run.unique())

    ind = df.model.isin(["hierarchical_bayesian"])
    hb = df[ind].reset_index(drop=True).copy()
    hb = hb.sort_values(by=["run"]).reset_index(drop=True).copy()
    hb = hb[metric].values

    ind = df.model.isin(["non_hierarchical_bayesian"])
    nhb = df[ind].reset_index(drop=True).copy()
    nhb = nhb.sort_values(by=["run"]).reset_index(drop=True).copy()
    nhb = nhb[metric].values

    return run, hb, nhb
```

```python
mae_mean = []
mse_mean = []
```

```python
src = "/home/vishu/repos/hbmep-paper/reports/down-sampled/1/combined/evaluation.csv"
down = pd.read_csv(src)

down.head()
```

```python
metric = "mae"

run, hb, nhb = get_scores(df=down, metric=metric)

sns.lineplot(x=run, y=hb, label="hb", marker="o")
sns.lineplot(x=run, y=nhb, label="nhb", marker="o")

t = stats.ttest_1samp(
    a=hb - nhb,
    popmean=0,
    alternative="less"
).pvalue

ranked = stats.wilcoxon(
    x=hb - nhb,
    alternative="less"
).pvalue

title = f"{metric}\nt-test={t:.02}, ranked-test={ranked:.02}"
plt.title(title)

plt.legend();
```

```python
mae_mean.append([hb.mean(), nhb.mean()])

print(hb.mean(), nhb.mean())
```

```python
metric = "mse"

run, hb, nhb = get_scores(df=down, metric=metric)

sns.lineplot(x=run, y=hb, label="hb", marker="o")
sns.lineplot(x=run, y=nhb, label="nhb", marker="o")

t = stats.ttest_1samp(
    a=hb - nhb,
    popmean=0,
    alternative="less"
).pvalue

ranked = stats.wilcoxon(
    x=hb - nhb,
    alternative="less"
).pvalue

title = f"{metric}\nt-test={t:.02}, ranked-test={ranked:.02}"
plt.title(title)

plt.legend();
```

```python
mse_mean.append([hb.mean(), nhb.mean()])

print(hb.mean(), nhb.mean())
```

```python
import os
import arviz as az

from numpyro.diagnostics import hpdi

from hbmep.model.utils import Site as site
```

```python
sorted(glob.glob(src))
```

```python
hb_hpdi = []
nhb_hpdi = []

row = []

for i in range(1, 11):
    src = f"/home/vishu/repos/hbmep-paper/reports/down-sampled/1/hb/run{i}/"

    a = az.from_netcdf(os.path.join(src, "hierarchical_bayesian/mcmc.nc"))
    a = a["posterior"][site.a].to_numpy()
    a = a.reshape(-1, 10, 3, 1)
    hpdi_interval = hpdi(a, prob=.95)
    len_ = (hpdi_interval[1, ...] - hpdi_interval[0, ...]).mean()
    row.append(len_)

hb_hpdi.append(row)


row = []

for i in range(1, 11):
    src = f"/home/vishu/repos/hbmep-paper/reports/down-sampled/1/nhb/run{i}/"

    a = az.from_netcdf(os.path.join(src, "non_hierarchical_bayesian/mcmc.nc"))
    a = a["posterior"][site.a].to_numpy()
    a = a.reshape(-1, 10, 3, 1)
    hpdi_interval = hpdi(a, prob=.95)
    len_ = (hpdi_interval[1, ...] - hpdi_interval[0, ...]).mean()
    row.append(len_)

nhb_hpdi.append(row)


for sparse_factor in [2, 3, 4, 5]:
    row = []

    for i in range(1, 11):
        src = f"/home/vishu/repos/hbmep-paper/reports/down-sampled/{sparse_factor}/run{i}/"

        a = az.from_netcdf(os.path.join(src, "hierarchical_bayesian/mcmc.nc"))
        a = a["posterior"][site.a].to_numpy()
        a = a.reshape(-1, 10, 3, 1)
        hpdi_interval = hpdi(a, prob=.95)
        len_ = (hpdi_interval[1, ...] - hpdi_interval[0, ...]).mean()
        row.append(len_)

    hb_hpdi.append(row)

    row = []

    for i in range(1, 11):
        src = f"/home/vishu/repos/hbmep-paper/reports/down-sampled/{sparse_factor}/run{i}/"

        a = az.from_netcdf(os.path.join(src, "non_hierarchical_bayesian/mcmc.nc"))
        a = a["posterior"][site.a].to_numpy()
        a = a.reshape(-1, 10, 3, 1)
        hpdi_interval = hpdi(a, prob=.95)
        len_ = (hpdi_interval[1, ...] - hpdi_interval[0, ...]).mean()
        row.append(len_)

    nhb_hpdi.append(row)
```
